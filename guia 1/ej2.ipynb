#### Recortar las letras

#### Recortar las estrellas y trasponerlas

```python
data = plt.imread("AFALogo.bmp")

#Se arma un copy para que el array sea modificable
copy_data=data.copy() 

plt.figure()
plt.imshow(data)
plt.title('imagen original')

#Se recorta la zona de las estrellas
stars_zone = copy_data[0:35,:]

plt.figure()
plt.imshow(stars_zone)
plt.title('zona de las estrellas')

#Se obtiene el color de las estrellas mirando su ubicación en los graficos
value = data[20,40]

#Se crea una mascara para las ubicaciones de las estrellas, considerando el color obtenido previamente
mask_stars = (stars_zone==value)

#Se modifica stars_zone con la máscara de estrellas, cambiandolas de color a 255
stars_zone[mask_stars] = 255

#Se pega la zona de las estrellas modificada en la copia del array de datos inicial
copy_data[0:35,:] = stars_zone

plt.figure()
plt.imshow(copy_data)
plt.title('imagen modificada')
```

#### Generar una máscara que separe el fondo del logotipo

```python
#Se crea una copia del array de datos original
data = plt.imread("AFALogo.bmp")
copy_data = data.copy()

#Se selecciona el color del fondo
value=copy_data [10,10]

#Se crea la máscara usando el color del fondo
mask =  (copy_data == value)

#Se cambia a 255 el valor del color del fondo usando la máscara
copy_data[mask] = 255

plt.figure()
plt.imshow(data)
plt.title('imagen original')

plt.figure()
plt.imshow(copy_data)
plt.title('imagen modificada')

```

#### Agregar la 3ra estrella

```python
#Se crea una copia del array de datos original
data = plt.imread("AFALogo.bmp")
copy_data = data.copy()

plt.figure()
plt.imshow(data)
plt.title('imagen original')

#Se obtiene el color de las estrellas mirando su ubicación en los graficos
value = data[20,40]
stars_zone = copy_data[0:35,:]

#Se recortan las estrellas
star1= np.copy(stars_zone[:,20:60])
star2= np.copy(stars_zone[:,60:110])

plt.figure()
plt.imshow(star1)
plt.title('Estrella 1')

plt.figure()
plt.imshow(star2)
plt.title('Estrella 2')

#Se pintan las 2 estrellas del mismo color que el fondo
stars_zone[:,:] = 0

#Se pinta todo el sector del mismo color que el fondo
plt.figure()
plt.imshow(stars_zone)
plt.title('eliminación de las estrellas')

#Se pegan las estrellas recortadas en los extremos del logo
stars_zone[:,0:40] = star1
stars_zone[:,77:127] = star2
stars_zone[:,43:83] = star1

#Se agregan las 3 estrellas
plt.figure()
plt.imshow(stars_zone)
plt.title('3 estrellas')

copy_data[0:35,:] = stars_zone
plt.figure()
plt.imshow(copy_data)
plt.title('Se agregan las 3 estrellas a la copia de la imagen original')


```
